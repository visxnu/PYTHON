```python
def triangle1(n):
  for i in range(1,n+1):
    for j in range(1,i+1):
      print("* ",end="")
    print()

triangle1(6)
```

**2**

```python
def triangle2(n):
  for i in range(n,0,-1):
    for j in range(1,i+1):
      print("* ",end="")
    print()

triangle2(6)
```

**3**

```python
def triangle3(n):
  m=1
  for i in range(1,n+1):
    for j in range(n-i):
      print(" ",end=" ")
    for k in range(1,2*i):
      print(m,end=" ")
    print()
    m=m+1

triangle3(6)
```
