Accessing Items

refering key

```python
dic={"name":"vishnu","age":21}
print(dic["name"])
```

get method

```python
dic={"name":"vishnu","age":21,"place":"kasaragod","mob num": 99888292}
x=dic.get("name")
print(x)
```

key method

```python
dic={"name":"vishnu","age":21,"place":"kasaragod","mob num": 99888292}
x=dic.keys()
print(x)
```

value method

```python
dic={"name":"vishnu","age":21,"place":"kasaragod","mob num": 99888292}
x=dic.values()
print(x)
```

item method

```python
dic={"name":"vishnu","age":21,"place":"kasaragod","mob num": 99888292}
x=dic.items()
print(x)
```

Add/Change dict item

```python
ClassE={"child1":"hanan","child2":"Edwin","child3":"afnan"}

ClassE["child4"]="saxon"
print(ClassE)
```

```python
ClassE={"child1":"hanan","child2":"Edwin","child3":"afnan"}

ClassE.update({"child4":"saxon"})
print(ClassE)
```

Remove methods

pop()

```python
a={"child1":"hanan","child2":"Edwin","child3":"afnan"}
a.pop("child1")
print(a)
```

popitem()

```python
a={"child1":"hanan","child2":"Edwin","child3":"afnan"}
a.popitem()
print(a)
```

del keyword

```python
a={"child1":"hanan","child2":"Edwin","child3":"afnan"}
del a["child1"]
print(a)
```

clear ()

```python
a={"child1":"hanan","child2":"Edwin","child3":"afnan"}
a.clear()
print(a)
```
